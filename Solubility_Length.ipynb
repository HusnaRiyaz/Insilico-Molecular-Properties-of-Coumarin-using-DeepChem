```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Predictedfile = pd.read_excel('Solubility_Pred.xlsx')
Smilespred=Predictedfile.iloc[:,1].values
Solpred = Predictedfile.iloc[:,2].values
```

```python
len_smiles=[]
for i in range(len(Smilespred)):
  len_smiles.append(len(Smilespred[i]))
```

```python
plt.scatter(len_smiles, Solpred, color = 'blue')
plt.title('Length of smiles vs predicted solubility')
plt.xlabel('length of smiles')
plt.ylabel('Predicted solubility')
plt.show()
```

```python
plt.scatter(len_smiles, Solpred, color = 'red')
plt.title('Length of smiles vs predicted solubility')
plt.xlabel('length of smiles')
plt.ylabel('Predicted solubility')
plt.show()
```
